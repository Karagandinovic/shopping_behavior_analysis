```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\Python Files\shopping_behavior_updated.csv', sep=',', encoding='utf-8')
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(r'[^\w]+', '_', regex=True)
)
df = df.set_index('customer_id')
df
```

```python
df = df.rename(columns={'purchase_amount_usd_': 'purchase_amount_usd'})
```

```python
df.info()
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df['frequency_of_purchases'] = df['frequency_of_purchases'].replace('Every 3 Months', 'Quarterly')
df['frequency_of_purchases'] = df['frequency_of_purchases'].replace('Bi-Weekly', 'Fortnightly')
df = df.drop('promo_code_used', axis=1)
df
```

```python
df.to_csv(r'C:\Users\user\Desktop\Python Files\shopping_behavior_updated.csv')
```

```python
from scipy import stats
```

```python
group_a = df[df['gender'] == 'Male']['purchase_amount_usd']
group_b = df[df['gender'] == 'Female']['purchase_amount_usd']

t_stat, p_value = stats.ttest_ind(group_a, group_b)

print('t-statistic: ', t_stat)
print('p-value: ', p_value)
```

```python
group_a = df[df['gender'] == 'Male']['previous_purchases']
group_b = df[df['gender'] == 'Female']['previous_purchases']

t_stat, p_value = stats.ttest_ind(group_a, group_b)

print('t-statistic: ', t_stat)
print('p-value: ', p_value)
```

```python
# creating age grous
df['age_group'] = (df['age'] // 10) * 10
df['age_group'] = df['age_group'].astype(str) + '-' + (df['age_group'] + 10).astype(str)

group1 = df[df['age_group'] == '10-20']['purchase_amount_usd']
group2 = df[df['age_group'] == '20-30']['purchase_amount_usd']
group3 = df[df['age_group'] == '30-40']['purchase_amount_usd']
group4 = df[df['age_group'] == '40-50']['purchase_amount_usd']
group5 = df[df['age_group'] == '50-60']['purchase_amount_usd']
group6 = df[df['age_group'] == '60-70']['purchase_amount_usd']
group7 = df[df['age_group'] == '70-80']['purchase_amount_usd']

anova = stats.f_oneway(group1, group2, group3, group4, group5, group6, group7)  # ANOVA
kruskal = stats.kruskal(group1, group2, group3, group4, group5, group6, group7)
print(anova)
print(kruskal)
```

```python
# ANOVA WELCH
import pingouin as pg

pg.welch_anova(dv='purchase_amount_usd', between='age_group', data=df)
```

```python
print(stats.shapiro(group1))  # тест на нормальность
print(stats.levene(group1, group2, group3, group4, group5, group6, group7))  # тест на одинаковость дисперсий
```

```python
# age_group vs previous_purchases

pg.welch_anova(dv='previous_purchases', between='age_group', data=df)
```

```python
group_a = df[df['subscription_status'] == 'Yes']['purchase_amount_usd']
group_b = df[df['subscription_status'] == 'No']['purchase_amount_usd']

t_stat, p_value = stats.ttest_ind(group_a, group_b)

print('t-statistic: ', t_stat)
print('p-value: ', p_value)
```

```python
group_a = df[df['subscription_status'] == 'Yes']['previous_purchases']
group_b = df[df['subscription_status'] == 'No']['previous_purchases']

t_stat, p_value = stats.ttest_ind(group_a, group_b)

print('t-statistic: ', t_stat)
print('p-value: ', p_value)
```

```python
print(pg.welch_anova(dv='purchase_amount_usd', between='frequency_of_purchases', data=df))
print(pg.welch_anova(dv='previous_purchases', between='frequency_of_purchases', data=df))
```

```python
# discount and purchase amount

group_a = df[df['discount_applied'] == 'Yes']['purchase_amount_usd']
group_b = df[df['discount_applied'] == 'No']['purchase_amount_usd']

t_stat, p_value = stats.ttest_ind(group_a, group_b)

print(t_stat, p_value)
```
